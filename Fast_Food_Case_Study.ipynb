```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('mcdonalds.csv')
df.head()
```

## Now we start with preliminary analysis of data and finding any discrepancies or null values in the data frame.
NOTE: We will also try to find unique data types and replace them with an alternative value for ease of calculation

```python
df.shape #rows,columns
```

```python
df.info() #column names, type, number of non null values
```

```python
#Performing basic EDA

age_ranges = [0, 10, 20, 30, 40, 50, 60, 70, 80]
df['Age Range'] = pd.cut(df['Age'], bins=age_ranges)
plt.figure(figsize=(10, 7))
sns.countplot(data=df, x='Age Range',color='#ccff66')
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.title('Count of Age Ranges')

plt.show()
```

We can see frequency of each age group

```python
gender_counts = df['Gender'].value_counts()
colors = ['#d966ff','#66b3ff']
labels = gender_counts.index
sizes = gender_counts.values
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%')
plt.title('Gender Distribution')

plt.show()
```
